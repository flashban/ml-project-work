```python
import pandas as pd

# Load the dataset
co2_emissions_df = pd.read_csv('../data/CO2 Emissions_Canada.csv')

# Display basic information about the dataset
print(co2_emissions_df.info())

# Display the first few rows
print(co2_emissions_df.head())
```

```python
# Display summary statistics
print(co2_emissions_df.describe())
```

```python
# Check for missing values
print(co2_emissions_df.isnull().sum())

# Handle missing values if any (example: fill with mean or drop)
# co2_emissions_df.fillna(co2_emissions_df.mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of CO2 Emissions
plt.figure(figsize=(10, 6))
sns.histplot(co2_emissions_df['CO2 Emissions(g/km)'], bins=30, kde=True)
plt.title('Distribution of CO2 Emissions')
plt.xlabel('CO2 Emissions (g/km)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot to explore relationships between variables
sns.pairplot(co2_emissions_df, vars=['Engine Size(L)', 'Cylinders', 'Fuel Consumption Comb (L/100 km)', 'CO2 Emissions(g/km)'])
plt.show()
```

```python
# Boxplot of CO2 Emissions by Vehicle Class
plt.figure(figsize=(12, 8))
sns.boxplot(x='Vehicle Class', y='CO2 Emissions(g/km)', data=co2_emissions_df)
plt.xticks(rotation=90)
plt.title('CO2 Emissions by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('CO2 Emissions (g/km)')
plt.show()
```
